```python
import mlrose_hiive
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import time
import timeit
```

# Define the problem: FlipFlop

```python
fitness = mlrose_hiive.FlipFlop()
problem1 = mlrose_hiive.DiscreteOpt(length=25, fitness_fn=fitness, maximize=True)
problem2 = mlrose_hiive.DiscreteOpt(length=50, fitness_fn=fitness, maximize=True)
problem3 = mlrose_hiive.DiscreteOpt(length=80, fitness_fn=fitness, maximize=True)
```

# Try Different Sizes

```python
def plot_fitness_curve(curve, problem, title):
  array_len = len(curve)
  x = np.arange(1, array_len+1, 1)
  y = curve[:,0]
  plt.title(title)
  plt.plot(x,y)
  plt.xlabel("Iteration")
  plt.ylabel("Fitness score")
  plt.show
```

## Random Hill Climb

```python
best_state_rhc1, best_fitness_rhc1, curve_rhc1 = mlrose_hiive.random_hill_climb(problem1, 
                                                                             #schedule=schedule, 
                                                                             max_iters=1000,
                                                                             curve=True)
title = "rhc iter vs fitness curve"
plot_fitness_curve(curve_rhc1,problem1,title)
```

```python
best_state_rhc2, best_fitness_rhc2, curve_rhc2 = mlrose_hiive.random_hill_climb(problem2, 
                                                                             #schedule=schedule, 
                                                                             max_iters=1000,
                                                                             curve=True)
title = "rhc iter vs fitness curve"
plot_fitness_curve(curve_rhc2,problem2,title)
```

```python
best_state_rhc3, best_fitness_rhc3, curve_rhc3 = mlrose_hiive.random_hill_climb(problem3, 
                                                                             #schedule=schedule, 
                                                                             max_iters=1000,
                                                                             curve=True)
title = "rhc iter vs fitness curve"
plot_fitness_curve(curve_rhc3,problem3,title)
```

```python
plt.figure()
x = np.arange(1, 21)
plt.plot(x,curve_rhc1[:,0][:20], label = "length 25")
plt.plot(x,curve_rhc2[:,0][:20], label = "length 50")
plt.plot(x,curve_rhc3[:,0][:20], label = "length 80")
plt.title("RHC iter vs fitness curve (different problem sizes)")
#plt.title('RHC')
plt.xlabel("Iteration")
plt.ylabel("fitness")
plt.legend()
plt.show
```
