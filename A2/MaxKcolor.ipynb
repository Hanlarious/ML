# Try Different Sizes

```python
def plot_fitness_curve(curve, problem, title):
  array_len = len(curve)
  x = np.arange(1, array_len+1, 1)
  y = curve[:,0] * -1
  #plt.title(problem + 'iter vs. loss curve')
  plt.title(title)
  plt.plot(x,y)
  plt.xlabel("Iteration")
  plt.ylabel("Fitness score")
  plt.show
```
